```python
import matplotlib.pyplot as plt
import numpy as np
# Data
models = ["RF", "RF tuned", "RF top 5 features", "SVM", "SVM tuned", "GB", 
          "GB tuned", "NN", "Bagging", "Bagging tuned", "Baseline: mean", "Baseline: LR"]
RMSE = [14.92, 14.92, 17.2, 21.67, 21.1, 21.18, 17.9, 20.93, 15.66, 15.02, 22.21, 21.95]
seconds = [72, 241, 24, 441, 36003, 21, 596, 110, 15, 172, 1, 1]

# Plot
fig, ax1 = plt.subplots(figsize=(14, 7))

color = 'tab:red'
ax1.set_xlabel('Model')
ax1.set_ylabel('Seconds (Including Tune Time)', color=color)
ax1.bar(models, seconds, color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_title('Comparison of Models, Time Taken and RMSE')
plt.xticks(rotation=45, ha='right')

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('RMSE', color=color)  
ax2.plot(models, RMSE, color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
